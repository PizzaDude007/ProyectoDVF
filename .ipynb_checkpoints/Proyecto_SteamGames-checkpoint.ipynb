```python
DatosFecha = Datos
DatosFecha['ReleaseDate'] = DatosFecha['ReleaseDate'].str.replace(' ','')
DatosFecha['ReleaseDate'] = DatosFecha['ReleaseDate'].str.replace('Jan','01')
DatosFecha['ReleaseDate'] = DatosFecha['ReleaseDate'].str.replace('Feb','02')
DatosFecha['ReleaseDate'] = DatosFecha['ReleaseDate'].str.replace('Mar','03')
DatosFecha['ReleaseDate'] = DatosFecha['ReleaseDate'].str.replace('Apr','04')
DatosFecha['ReleaseDate'] = DatosFecha['ReleaseDate'].str.replace('May','05')
DatosFecha['ReleaseDate'] = DatosFecha['ReleaseDate'].str.replace('Jun','06')
DatosFecha['ReleaseDate'] = DatosFecha['ReleaseDate'].str.replace('Jul','07')
DatosFecha['ReleaseDate'] = DatosFecha['ReleaseDate'].str.replace('Aug','08')
DatosFecha['ReleaseDate'] = DatosFecha['ReleaseDate'].str.replace('Sep','09')
DatosFecha['ReleaseDate'] = DatosFecha['ReleaseDate'].str.replace('Oct','10')
DatosFecha['ReleaseDate'] = DatosFecha['ReleaseDate'].str.replace('Nov','11')
DatosFecha['ReleaseDate'] = DatosFecha['ReleaseDate'].str.replace('Dec','12')
DatosFecha['ReleaseDate'] = DatosFecha['ReleaseDate'].str.replace('[aA-zZ]','', regex=True)
DatosFecha = DatosFecha.dropna()

#DatosFecha['ReleaseDate'] = DatosFecha['ReleaseDate'].astype('int')

DatosFecha
```

```python
JoinData = pd.DataFrame()

for column in Datos.loc[:,'GenreIsIndie':'GenreIsMassivelyMultiplayer'].columns:
    #DatosOne = Datos[Datos[column].eq(1)].drop_duplicates()
    JoinData = JoinData.append(Datos.groupby(column).agg({'Metacritic':['mean','min','max','count'],
                                                          'RecommendationCount':['mean','min','max','sum'],
                                                          'PriceInitial':['mean','min','max','sum'],
                                                         }).loc[1:1], ignore_index=True)
    
JoinData.insert(0, 'Genre', Datos.loc[:,'GenreIsIndie':'GenreIsMassivelyMultiplayer'].columns)
JoinData['Genre'] = JoinData['Genre'].str.replace('GenreIs','')

    
JoinData
```

```python
plt.figure(figsize=(20, 5))
plt.bar(JoinData['Genre'],JoinData[('PriceInitial','mean')], color='green', label='Precio')
plt.xlabel('Género')
plt.ylabel('Videojuego')
plt.title('Steam Games')
plt.grid(True)
plt.legend()
plt.show()
```
